# Make heatmaps by clusters

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'data_subsets/DIP_drp.csv', index_col=0)
df
```

```python
LNV_df = df[(df['Idents'] == '2:s_LNv') & (df['exp_time'] == 'ZT14')]\
    .drop(['Idents', 'date', 'experiment', 'exp_time'], axis='columns')
LNV_df
```

```python
sns.heatmap(LNV_df.T, 
            square=True, 
            cmap='viridis', 
            cbar=True, 
            vmin=-3, 
            vmax=3, 
            center=0, 
            yticklabels=True, 
            xticklabels=False)
```
